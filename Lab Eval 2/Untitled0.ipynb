```python
observed_statistic = abs ( 100 * (755 / 929) - 75)
observed_statistic
```

## 4

```python
from datascience import *
mendel_proportions = make_array(0.75, 0.25)
mendel_proportion_purple = mendel_proportions.item(0)
sample_size = 929
```

```python
def one_simulated_distance():
    sample_proportion_purple = sample_proportions(929, mendel_proportions).item(0)
    return 100 * abs(sample_proportion_purple - mendel_proportion_purple)
```

```python

repetitions = 10000
distances = make_array()
for i in np.arange(repetitions):
    distances = np.append(distances, one_simulated_distance())
```

# New Section

```python
Table().with_column('Distance between Sample % and 75%', distances).hist()
plt.ylim(-0.02, 0.5)
plt.title('Prediction Made by the Null Hypothesis')
plt.scatter(observed_statistic, 0, color='red', s=40);
```

From this, again we accept the null hypothesis and can conclude mendel's model works

```python
import numpy as np
import numpy.random as npr
import scipy.stats as st
```

```python
n=30
xbar=65.10
mu=60
sigma=13.46
alpha=0.05
```

```python
zcritical=st.norm.ppf(alpha/2)
zcritical
```

```python
z=(xbar-mu)/(sigma/np.sqrt(n))
z
```

```python
if (z>zcritical and z<abs(zcritical)):
  print('Null hypothesis cannot be rejected')
else:
  print('Null hypothesis is rejected')
```

As our test statistic value is lying in the critical region, we reject the null hypothesis
